### Step 1. Import the necessary libraries:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv).

```python
url = "https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv"
```

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv(url)
tips
```

### Step 4. Delete the Unnamed 0 column

```python
tips.drop(columns=['Unnamed: 0'], inplace=True)
```

### Step 5. Plot the total_bill column histogram

```python
sns.histplot(tips['total_bill'], kde=True)
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Histogram of Total Bill')
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatterplot of Total Bill and Tip')
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
sns.pairplot(tips)
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
sns.boxplot(x='day', y='total_bill', data=tips)
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Total Bill by Day')
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
sns.scatterplot(x='tip', y='day', hue='sex', data=tips)
plt.xlabel('Tip')
plt.ylabel('Day')
plt.title('Scatterplot of Tip by Day (Distinguished by Sex)')
plt.legend(title='Sex')
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Total Bill by Day and Time')
plt.legend(title='Time')
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(tips[tips['time'] == 'Dinner']['tip'], kde=True)
plt.xlabel('Tip')
plt.ylabel('Frequency')
plt.title('Dinner Tip Histogram')

plt.subplot(1, 2, 2)
sns.histplot(tips[tips['time'] == 'Lunch']['tip'], kde=True)
plt.xlabel('Tip')
plt.ylabel('Frequency')
plt.title('Lunch Tip Histogram')

plt.tight_layout()
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=tips[tips['sex'] == 'Male'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Male - Total Bill vs Tip (Smoker vs Non-Smoker)')

plt.subplot(1, 2, 2)
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=tips[tips['sex'] == 'Female'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Female - Total Bill vs Tip (Smoker vs Non-Smoker)')

plt.tight_layout()
plt.show()
```

### BONUS: Create your own question and answer it using a graph.

```python
# Is there any relationship between the total bill and the tip amount, considering the day of the week and whether the customer is a smoker or not?
plt.figure(figsize=(10, 6))
sns.scatterplot(x='total_bill', y='tip', hue='day', style='smoker', data=tips)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Total Bill vs Tip (Day and Smoker Status)')
plt.legend(title='Day', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
