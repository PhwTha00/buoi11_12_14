### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv).

```python
url = "https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')
```

### Step 3. Assign it to a variable called chipo.

```python
chipo.head()
```

### Step 4. See the first 10 entries

```python
chipo.head(10)
```

### Step 5. Create a histogram of the top 5 items bought

```python
top_items = chipo['item_name'].value_counts().head(5)

plt.bar(top_items.index, top_items.values)
plt.xlabel('Items')
plt.ylabel('Number of Orders')
plt.title('Top 5 Items Bought at Chipotle')
plt.xticks(rotation=45)
plt.show()
```

### Step 6. Create a scatterplot with the number of items orderered per order price
#### Hint: Price should be in the X-axis and Items ordered in the Y-axis

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
orders = chipo.groupby('order_id').agg({'quantity': 'sum', 'item_price': 'sum'})

plt.scatter(orders['item_price'], orders['quantity'])
plt.xlabel('Order Price ($)')
plt.ylabel('Items Ordered')
plt.title('Number of Items Ordered per Order Price')
plt.show()
```

### Step 7. BONUS: Create a question and a graph to answer your own question.

```python
#Question: What is the distribution of order prices at Chipotle?
plt.hist(orders['item_price'], bins=20)
plt.xlabel('Order Price ($)')
plt.ylabel('Frequency')
plt.title('Distribution of Order Prices at Chipotle')
plt.show()
```
