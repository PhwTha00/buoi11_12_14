# Visualizing the Titanic Disaster

### Introduction:

This exercise is based on the titanic Disaster dataset avaiable at [Kaggle](https://www.kaggle.com/c/titanic).  
To know more about the variables check [here](https://www.kaggle.com/c/titanic/data)


### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv)

### Step 3. Assign it to a variable titanic

```python
titanic = pd.read_csv("https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic.csv")
print(titanic)
```

### Step 4. Set PassengerId as the index

```python
titanic.set_index('PassengerId', inplace=True)
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
plt.figure(figsize=(8, 6))
titanic['Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightcoral'])
plt.title('Male/Female Proportion in Titanic Dataset')
plt.ylabel('')
plt.show()
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Fare', hue='Sex', data=titanic)
plt.title('Fare Paid vs Age (Colored by Gender)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend(title='Gender')
plt.show()
```

### Step 7. How many people survived?

```python
survived_count = titanic['Survived'].sum()
print("Number of people who survived:", survived_count)
```

### Step 8. Create a histogram with the Fare payed

```python
plt.figure(figsize=(10, 6))
sns.histplot(titanic['Fare'], bins=20, kde=True, color='skyblue')
plt.title('Histogram of Fare Paid')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

### BONUS: Create your own question and answer it.

```python
# What was the survival rate among different passenger classes?
survival_rate_by_class = titanic.groupby('Pclass')['Survived'].mean() * 100

plt.figure(figsize=(8, 6))
sns.barplot(x=survival_rate_by_class.index, y=survival_rate_by_class.values, palette='Set2')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate (%)')
plt.xticks(ticks=[0, 1, 2], labels=['1st Class', '2nd Class', '3rd Class'])
plt.show()

print("Survival rate by passenger class:")
print(survival_rate_by_class)
```
